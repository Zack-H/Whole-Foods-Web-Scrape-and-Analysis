```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sqlalchemy import create_engine
```

```python
url = 'https://www.wholefoodsmarket.com/products/all-products?featured=on-sale'
```

```python
request = requests.get(url)
```

```python
request
```

```python
request.text
```

```python
soup = BeautifulSoup(request.text, 'html.parser')
```

```python
print(soup.prettify)
```

```python
products = soup.findAll('div', attrs ={'class':'w-pie--product-tile__content'})
```

```python
products
```

```python
productDetails = {
    'name':[],
    'category':[],
    'originalPrice':[],
    'salePrice':[],
    'primePrice':[],
    'url':[]
}

for product in products:
    name = product.find('h3', attrs = {'class':'w-cms--font-body__sans-bold'}).text
    productDetails['name'].append(name)
    print('Name: ', name)
    
    category = product.find('span', attrs = {'class':'w-cms--font-disclaimer'}).text
    productDetails['category'].append(category)
    print('Category: ', category)
    
    originalPrice = product.find('span', attrs = {'class':'regular_price has_sale'}).text
    productDetails['originalPrice'].append(originalPrice)
    print('OriginalPrice: ', originalPrice)
    
    salePrice = product.find('span', attrs = {'class':'sale_price'}).text
    productDetails['salePrice'].append(salePrice)
    print('SalePrice: ', salePrice)
    
    primePrice = product.find('span', attrs = {'class':'prime_price prime_incremental'}).text
    productDetails['primePrice'].append(primePrice)
    print('PrimePrice: ', primePrice)
    
    url = product.find('a', attrs = {'class':'w-pie--product-tile__link'})['href']
    productDetails['url'].append(url)
    print('Url: ', url)
    
    print('-'*75)
```

# Attempt 3
Go up a level to see if the 'a' tag can't be the outer most tag

```python
products = soup.findAll('div', attrs ={'class':'w-pie--product-tile'})
```

```python
productDetails = {
    'name':[],
    'category':[],
    'originalPrice':[],
    'salePrice':[],
    'primePrice':[],
    'link':[]
}

for product in products:
    name = product.find('h3', attrs = {'class':'w-cms--font-body__sans-bold'}).text
    productDetails['name'].append(name)
    print('Name: ', name)
    
    category = product.find('span', attrs = {'class':'w-cms--font-disclaimer', 'data-testid':'product-tile-brand'}).text
    productDetails['category'].append(category)
    print('Category: ', category)
    
    originalPrice = product.find('span', attrs = {'class':'regular_price has_sale'}).text[1:]
    originalPricePartitioned = originalPrice.partition('/')
    originalPriceNoUnits = originalPricePartitioned[0]
    productDetails['originalPrice'].append(originalPriceNoUnits)
    print('OriginalPrice: ', originalPriceNoUnits)
    
    salePrice = (product.find('span', attrs = {'class':'sale_price'}).text)[10:]
    salePricePartitioned = salePrice.partition('/')
    salePriceNoUnits = salePricePartitioned[0]
    #parse '2 for x' strings
    if salePriceNoUnits.find('for') != -1:
        quantity = int(salePriceNoUnits[0])
        print('quantity: ', quantity)
        batchPrice = float(salePriceNoUnits[7:])
        print('batch price: ', batchPrice)
        finalSalePriceNoUnits = batchPrice / quantity
    else:
        finalSalePriceNoUnits = salePriceNoUnits[1:]
    print('finalSalePriceNoUnits: ',finalSalePriceNoUnits)
    productDetails['salePrice'].append(str(finalSalePriceNoUnits))
    print('SalePrice: ', finalSalePriceNoUnits)
    
    try:
        primePrice = (product.find('span', attrs = {'class':'prime_price prime_incremental'}).text)[18:]
        primePricePartitioned = primePrice.partition('/')
        primePriceNoUnits = primePricePartitioned[0]
        #parse '2 for x' strings
        if primePriceNoUnits.find('for') != -1:
            quant = int(primePriceNoUnits[0])
            print('quant: ', quant)
            batch = float(primePriceNoUnits[7:])
            print('batch: ',batch)
            finalPrimePriceNoUnits = batch / quant
        else:
            finalPrimePriceNoUnits = primePriceNoUnits[1:]
        print('finalPrimePriceNoUnits: ',finalPrimePriceNoUnits)
        productDetails['primePrice'].append(str(finalPrimePriceNoUnits))
        print('PrimePrice: ', finalPrimePriceNoUnits)
    except:
        print('NO PRIME OPTION')
        productDetails['primePrice'].append(None)
    
    link = product.find('a', attrs = {'class':'w-pie--product-tile__link', 'data-testid':'product-tile-link'})['href']
    productDetails['link'].append(link)
    print('Link: ', link)
    
    print('-'*75)
```
